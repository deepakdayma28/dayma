```python
from sklearn.preprocessing import LabelEncoder
```

```python
POSTED_BY=LabelEncoder()
UNDER_CONSTRUCTION=LabelEncoder()
RERA=LabelEncoder()
BHK_NO=LabelEncoder()
BHK_OR_RK=LabelEncoder()
SQUARE_FT=LabelEncoder()
READY_TO_MOVE=LabelEncoder()
RESALE=LabelEncoder()
ADDRESS=LabelEncoder()
LONGITUDE=LabelEncoder()
LONGITUDE1=LabelEncoder()
TARGETPRICE=LabelEncoder()
```

```python
house["POSTED_BY"]=POSTED_BY.fit_transform(house["POSTED_BY"])
house["UNDER_CONSTRUCTION"]=POSTED_BY.fit_transform(house["UNDER_CONSTRUCTION"])
house["RERA"]=POSTED_BY.fit_transform(house["RERA"])
house["BHK_NO."]=POSTED_BY.fit_transform(house["BHK_NO."])
house["BHK_OR_RK"]=POSTED_BY.fit_transform(house["BHK_OR_RK"])
house["SQUARE_FT"]=POSTED_BY.fit_transform(house["SQUARE_FT"])
house["READY_TO_MOVE"]=POSTED_BY.fit_transform(house["READY_TO_MOVE"])
house["RESALE"]=POSTED_BY.fit_transform(house["RESALE"])
house["ADDRESS"]=POSTED_BY.fit_transform(house["ADDRESS"])
house["LONGITUDE"]=POSTED_BY.fit_transform(house["LONGITUDE"])
house["LONGITUDE.1"]=POSTED_BY.fit_transform(house["LONGITUDE.1"])
house["TARGET PRICE"]=POSTED_BY.fit_transform(house["TARGET PRICE"])
```
